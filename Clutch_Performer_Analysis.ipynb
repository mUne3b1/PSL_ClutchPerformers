```python
import pandas as pd

# Load the dataset
file_path = 'C:/Users/92324/Desktop/PSL/PSL_Season_09_dataset.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataset
print(data.head())
```

```python
# Fill NaN values in the extras columns with 0
data.fillna({'wides': 0, 'noballs': 0, 'byes': 0, 'legbyes': 0, 'penalty': 0}, inplace=True)

# Display the first few rows to check the changes
print(data.head())
```

```python
# Extract over information from the ball column
data['over'] = data['ball'].apply(lambda x: int(x))

# Display the first few rows to check the new column
print(data.head())
```

```python
# Calculate runs off bat and total runs
data['runs_off_bat'] = data['runs_off_bat'].fillna(0)
data['total_runs'] = data['runs_off_bat'] + data['extras']

# Display the first few rows to check the new columns
print(data.head())
```

```python
# Filter data for overs 16-20
death_overs_data = data[(data['over'] >= 16) & (data['over'] <= 20)]

# Display the filtered data
print(death_overs_data.head())
```

```python
# Group by match_id, innings, over, batting_team, bowling_team, and striker to get cumulative runs, balls faced, and wickets in death overs
death_overs_summary = death_overs_data.groupby(['match_id', 'innings', 'over', 'batting_team', 'bowling_team', 'striker']).agg(
    total_runs=('total_runs', 'sum'),
    balls_faced=('ball', 'count'),
    wickets=('wicket_type', 'count')
).reset_index()

# Calculate strike rate
death_overs_summary['strike_rate'] = (death_overs_summary['total_runs'] / death_overs_summary['balls_faced']) * 100

# Display the death overs summary
print(death_overs_summary.head())
```

```python
# Group by striker to get overall performance in death overs
clutch_performers = death_overs_summary.groupby('striker').agg(
    total_runs=('total_runs', 'sum'),
    balls_faced=('balls_faced', 'sum'),
    wickets=('wickets', 'sum'),
    average_strike_rate=('strike_rate', 'mean')
).reset_index()

# Filter for top performers based on total runs and strike rate
top_clutch_performers = clutch_performers[(clutch_performers['total_runs'] > 65) & (clutch_performers['average_strike_rate'] > 150)]

# Display the top clutch performers
print(top_clutch_performers.head(10))  # Display the top 10 performers
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the theme for seaborn
sns.set_theme(style="whitegrid")

# Group by batting_team and over to get average runs per over in death overs
team_performance = death_overs_summary.groupby(['batting_team', 'over']).agg(
    avg_runs_per_over=('total_runs', 'mean'),
    avg_wickets_per_over=('wickets', 'mean')
).reset_index()

# Create a bar plot for average runs per over in death overs by team
plt.figure(figsize=(14, 8))
sns.barplot(data=team_performance, x='over', y='avg_runs_per_over', hue='batting_team', ci=None)
plt.title('Average Runs Per Over in Death Overs by Team')
plt.xlabel('Over')
plt.ylabel('Average Runs Per Over')
plt.legend(title='Team', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# Create a bar plot for top clutch performers in death overs
plt.figure(figsize=(14, 8))
sns.barplot(data=top_clutch_performers, x='striker', y='total_runs', ci=None)
plt.title('Top Clutch Performers in Death Overs')
plt.xlabel('Player')
plt.ylabel('Total Runs in Death Overs')
plt.xticks(rotation=90)
plt.show()
```

```python

# Identify the top performer for each team based on total runs and strike rate
team_clutch_performers['performance_score'] = team_clutch_performers['total_runs'] * team_clutch_performers['average_strike_rate']
top_team_performers = team_clutch_performers.loc[team_clutch_performers.groupby('batting_team')['performance_score'].idxmax()]

# Display the top performers for each team
print(top_team_performers)

# Plot the top performers for each team
plt.figure(figsize=(14, 8))
sns.barplot(data=top_team_performers, x='batting_team', y='total_runs', hue='striker', ci=None)
plt.title('Top Clutch Performers in Death Overs by Team')
plt.xlabel('Team')
plt.ylabel('Total Runs in Death Overs')
plt.legend(title='Player', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
